```python
unemployment_condensed = {}
with open('STLABOUR_03012021061646979.csv') as csv_file2:
    unemployment_data = csv.reader(csv_file2, delimiter = ",")
    for line in unemployment_data:
        
        # big conditions to to eliminate data from countries/areas not present in COVID data
        if line[8][:4] == "2020" and "Euro" not in line[1] and "Total" not in line[1] and "7" not in line[1] and line[1] != "Latvia" and line[1] != "Lithuania":
            
            # adds each country
            if line[1] not in unemployment_condensed.keys():
                unemployment_condensed[line[1]] = {}
                
            # adds the unemployment rate
            unemployment_condensed[line[1]][line[8]] = float(line[16])
            
# creates countries list to be used later and further cleans data
countries = list(unemployment_condensed.keys())
countries.remove("Korea")
```

```python
import csv

condensed = {}
with open('OxCGRT_latest1.csv') as csvfile:
    covid_data = csv.reader(csvfile, delimiter = ",")
    for line in covid_data:
        
        # valid country
        if line[0] in countries and line[2] == "":
            
            # adds country to dictionary if it is new
            if line[0] not in condensed.keys():
                condensed[line[0]] = {}
                
            # only gets data from the first of the month
            if str(line[5][-2:]) == "01" and str(line[5][3]) == "0": #first of the month
                cases = line[37]
                deaths = line[38]
                
                # deals with blacks
                if cases == "":
                    cases = 0
                if deaths == "":
                    deaths = 0
                    
                # makes the date the same form as in unemployment data
                date = line[5][:4] + "-" + line[5][4:-2]
                
                # adds the data to the line
                condensed[line[0]][date] = {"cases":int(cases), "deaths":int(deaths)}



#calculates new cases and deaths total
new_cases_metric = {}
percent_of_last_month_cases_metric = {}
new_deaths_metric = {}
percent_of_last_month_deaths_metric = {}


for country, data in condensed.items():
    
    # adds each country
    new_cases_metric[country] = {}
    percent_of_last_month_cases_metric[country] = {}
    new_deaths_metric[country] = {}
    percent_of_last_month_deaths_metric[country] = {}
    
    total_cases = 0
    cases_last_month = 0
    total_deaths = 0
    deaths_last_month = 0
    
    # for each month
    for date, info in data.items():
        
        
        # calculates new cases and deaths
        new_cases = info["cases"] - total_cases
        new_deaths = info["deaths"] - total_deaths
        if new_cases < 0:
            new_cases = 0
        if new_deaths < 0:
            new_deaths = 0
        
        # adds them to metric dictionaries
        new_cases_metric[country][date] = new_cases
        new_deaths_metric[country][date] = new_deaths
        
        # adds percent cases and deaths metrics to their respective dictionaries
        if cases_last_month == 0:
            percent_of_last_month_cases_metric[country][date] = 0
        else:
            percent_of_last_month_cases_metric[country][date] = new_cases/cases_last_month*100
        if deaths_last_month == 0:
            percent_of_last_month_deaths_metric[country][date] = 0
        else:
            percent_of_last_month_deaths_metric[country][date] = new_deaths/deaths_last_month*100
           
        # updates carryover variables for next loop
        total_cases = info["cases"]
        total_deaths = info["deaths"]
        cases_last_month = new_cases
        deaths_last_month = new_deaths
```

```python
# makes csv files

# one for each country
for country in countries:
    with open(f"{country}1.csv", 'w') as file:
        file_writer = csv.writer(file, delimiter = ",")
        
        # column headers
        file_writer.writerow(["Date", "New Cases", "New Deaths", "% Last Month Cases", "% Last Month Deaths","Unemployment Rate"])
        
        # adds data from unemployment rate and metric dictionaries
        for date, rate in unemployment_condensed[country].items():
            file_writer.writerow([date, new_cases_metric[country][date], new_deaths_metric[country][date], percent_of_last_month_cases_metric[country][date], percent_of_last_month_deaths_metric[country][date], rate])
```
